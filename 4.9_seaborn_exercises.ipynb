```python
# use iris database 
from pydataset import data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# convert to pandas dataset and preview data
iris = data('iris')
iris = pd.DataFrame(iris)
iris
```

```python
# What does the distribution of petal lengths look like?
sns.distplot(iris['Petal.Length'])
plt.show()
plt.close()
```

```python
# Is there a correlation between petal length and petal width?

from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(iris["Petal.Length"],iris["Petal.Width"])

ax = sns.regplot(x = "Petal.Length", y = "Petal.Width", data = iris, color='b', 
                 line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)})

ax.legend() # plot legend
plt.show()
plt.close()

# sns.regplot(x = "Petal.Length", y = "Petal.Width", data = iris, corr_func=stats.pearsonr)
```

```python
sns.relplot(x = "Petal.Length", y = "Petal.Width", data = iris)
plt.text(5,
         iris["Petal.Length"].corr(iris["Petal.Width"]), # s
         'Correlation Coefficient: {:.2f}'.format(iris["Petal.Length"].corr(iris["Petal.Width"])),
        fontsize = 16) # annotate correlation 
plt.show()
plt.close()
```

```python
# Using the chipotle dataset from the previous exercise, 
# create a bar chart that shows the 4 most popular items and the revenue produced by each.

# Getting data from SQL databases
# Create a function, get_db_url. 
# accept a username, hostname, password, and database name 
# return a url to for accessing database in SQL server

def get_db_url():
    url = f'mysql+pymysql://{user}:{password}@{host}/{database_name}'
    return url
from env import host, user, password
database_name = input("Input desired database name: ")
query = input("Key in the query")
chipotle = pd.read_sql(query,get_db_url())
chipotle.head(5)
```

```python
chipotle["item_price"] = chipotle.item_price.str.strip("$")
chipotle
```

```python
chipotle["item_price"] = chipotle.item_price.astype(float)
```

```python
chipotle.dtypes
```

```python
# create a bar chart that shows the 4 most popular items and the revenue produced by each.
vis_data_4_popular_revenue = chipotle[["item_name","quantity","item_price"]]
vis_data_4_popular_revenue
```

```python
top_4 = vis_data_4_popular_revenue.groupby(['item_name','item_price'], as_index=False).quantity.count()
top_4
```

```python
top_4.sort_values(by = 'quantity',ascending = False)
```

```python
top_4['item_name'] = top_4["item_name"].str.replace(' ','_')
top_4
```

```python
top_4 = top_4.reset_index()
top_4
```

```python
popular_4 = (chipotle.groupby("item_name")
 .sum()
.drop(columns='order_id')
.rename(columns={'quantity':'unit_sold','item_price':'revenue'})
.sort_values(by = 'unit_sold', ascending = False)
.head(4)
.reset_index())

popular_4
```
