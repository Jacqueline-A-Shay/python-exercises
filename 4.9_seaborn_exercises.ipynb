```python
# use iris database 
from pydataset import data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# convert to pandas dataset and preview data
iris = data('iris')
iris = pd.DataFrame(iris)
iris
```

```python
# What does the distribution of petal lengths look like?
sns.distplot(iris['Petal.Length'])
plt.show()
plt.close()
```

```python
# Is there a correlation between petal length and petal width?

from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(iris["Petal.Length"],iris["Petal.Width"])

ax = sns.regplot(x = "Petal.Length", y = "Petal.Width", data = iris, color='b', 
                 line_kws={'label':"y={0:.1f}x+{1:.1f}".format(slope,intercept)})

ax.legend() # plot legend
plt.show()
plt.close()

# sns.regplot(x = "Petal.Length", y = "Petal.Width", data = iris, corr_func=stats.pearsonr)
```

```python
sns.relplot(x = "Petal.Length", y = "Petal.Width", data = iris)
plt.text(5,
         iris["Petal.Length"].corr(iris["Petal.Width"]), # s
         'Correlation Coefficient: {:.2f}'.format(iris["Petal.Length"].corr(iris["Petal.Width"])),
        fontsize = 16) # annotate correlation 
plt.show()
plt.close()
```
