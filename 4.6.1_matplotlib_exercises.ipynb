```python
x = range(500)
y = [math.sqrt(i) for i in x]
plt.subplot(331)
plt.plot(x,y,c = 'blue')
plt.title('$y = x$')

x = range(-100,100)
y = [i**3 for i in x]
plt.subplot(333)
plt.plot(x,y, c = 'red')
plt.title('$y = x^3$')


x = range(-500,500)
y =[math.tan(i) for i in x]
plt.subplot(337)
plt.plot(x,y, c = "green")
plt.title("$y = tan (x)$")

x = range(-100,303)
y = [2*i for i in x]
plt.subplot(339)
plt.plot(x,y, c = "orange")
plt.title("$y = 2^x$")

plt.suptitle('Math Demo')

plt.show()
```
